## 데이터 처리 연습 : 서울 따릉이 데이터

### to_datetime

```python
df = pd.read_csv('data/seoul_bicycle.csv')
df
```

```python
df.info()
```

```python
df1 = df.copy()
```

```python
df1['대여일자'].unique()
```

```python
df1['대여일자'] = pd.to_datetime(df1['대여일자'],format='%b-%d-%Y') #데이터 타입이라는 객체로 변경
```

### to_numeric

```python
df1.info()
```

```python
df1['운동량'] = pd.to_numeric(df1['운동량'], errors='coerce')
```

```python
df1.info()
```

```python
df1.tail()
```

```python
bins = [0, 1000, 2000, df1['운동량'].max()]
```

```python
pd.cut(df1['운동량'], bins).value_counts()
```

```python
df1.describe()
```

```python
df1['운동량_cut'] = pd.cut(df1['운동량'], bins=10)
df1['운동량_cut'].value_counts()
```

```python
df1['운동량_qcut'] = pd.qcut(df1['운동량'], q=10)
df1['운동량_qcut'].value_counts()
```

```python
qcut_bins = [0, 0.2, 0.8, 1]
pd.qcut(df1['운동량'], qcut_bins)
```

```python
df1['운동량'] = pd.qcut(df1['운동량'], qcut_bins, labels=['적음', '보통', '많음'])
```

```python
df1.loc[df1['운동량'] == '보통']
```

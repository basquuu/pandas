### fillna

```python
df = sns.load_dataset('titanic')
```

```python
df1 = df.copy()
```

```python
df1['age'].fillna(100).tail()
```

```python
df1['age'] = df1['age'].fillna(100)
```

```python
df['deck'].fillna('C')
```

```python
#df['deck'].fillna('No Data')
# => 카테고리 타입은 카테고리에 없는 정보는 넣을 수 없음
```

```python
df1['deck'].cat.add_categories('No Data')
df1['deck'].fillna()
```

```python
df1 = df.copy()
```

```python
mean_age = df1['age'].mean()
df1['age'].fillna(mean_age).tail()
```

```python
median_age = df1['age'].median()
df1['age'].fillna(mean_age).tail()
```

- 남자 승객의 결측치 => 남자 승객의 평균 채우기 
- 여자 승객의 결측치 => 여자 승객의 평균 채우기

```python
cond1 = df1['age'].isnull()
```

```python
male = df1['sex'] == 'male'
female = df1['sex'] == 'female'
male_age = df1['age'].loc[male].mean()
female_age = df1['age'].loc[female].mean()
```

```python
#cond1 = df['age'].isnull()
df.loc[cond1,'age'] = 30
```

```python
df1.loc[cond1 & male, 'age'] = male_age
df1.loc[cond1 & female, 'age'] = female_age
```

```python
df1['age'].mean()
```

```python
df1['deck'].mode()
```

```python
df1['deck'].fillna(df1['deck'].mode()[0])
```

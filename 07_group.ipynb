```python
import numpy as np
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
```

```python
df
```

## apply

```python
df['who'].value_counts()
```

```python
def trans_who(x):
    if x == 'man':
        return '남자'
    elif x == 'woman':
        return '여자'
    else:
        return '아이'
```

```python
df['who'].apply(trans_who)
```

```python
def fare_age(x):
    return x['fare'] / x['age']
```

```python
df.apply(fare_age, axis=1)
```

```python
df['survived']
```

## groupby
- 데이터를 바라볼 때 특정 데이터를 기준으로 grouping 할 때 사용

```python
df.groupby('sex').mean(numeric_only=True)
```

```python
df.groupby(['sex','pclass']).mean(numeric_only=True)
```

```python
df.groupby(['sex','pclass'])['survived'].mean(numeric_only=True)
```

```python
df.groupby(['sex','pclass']).mean(numeric_only=True)[['survived','fare']] 
```

## reset_index

```python
df.groupby(['sex','pclass'])['survived'].mean().reset_index() #index가 추가됨 
```

```python
df.groupby(['sex','pclass'])['survived', 'age'].agg(['mean','sum'])
```

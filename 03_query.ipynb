## loc - 조건필터링

```python
cond = df['who'] == 'man'
cond
```

- case 1 : df[cond]

```python
df[cond].head()
```

- cas2: df.loc[cond]

```python
df.loc[cond].head()
```

```python
df[cond]['age'] == 100
```

```python
df.loc[cond, 'age'] = 100
```

```python
cond1 = df['fare'] > 30
cond2 = df['who'] == 'woman'
```

```python
df.loc[cond1 & cond2].head()
```

```python
df.loc[cond1 | cond2].head()
```

```python
cond = df['fare'] < 30
df['fare'].where(cond, 1).tail() #특정 상황을 만족하지 못할 경우 값을 대체하는 것임 
```
